Машиное обучение позволяет найти правила по которым  реплика введеная пользователем  будет соотнесена с интентом датасета

```
x_text=[]
y=[]
#выделяем из датасета examples в виде словаря
for intent,intent_data in BOT_CONFIG['intents'].items():
      for examples in intent_data['examples']:
        x_text.append(examples)
        y.append(intent)
#Векторное представление текста в виде чисел        
vectorizer = CountVectorizer(analyzer='char',ngram_range=(2,4))
X = vectorizer.fit_transform(x_text)
#Запускаем на обучение классификатор
clf =LinearSVC ()
clf.fit(X, y)

```

```
def classify_intent(replica):
    replica=clear_phrase(replica)
    intent=clf.predict(vectorizer.transform([replica]))[0] #Запускаем на предсказание в классификатор векторизированную реплику
    for example in BOT_CONFIG['intents'][intent]['examples']:
        example=clear_phrase(example)
        distance=nltk.edit_distance(replica, example)
        if example and distance/len( example)<0.5:    #Защита от случайных фраз
           return  intent
```

Проверем на вхождение каждого символа в словарь alfabet

```
def clear_phrase(phrase):
   phrase=phrase.lower()
   alfabet='абвгдеёжзийклмнопрстуфцчшщэюя_ '
   result=''.join(symbol for symbol in phrase if symbol in alfabet)
   return result.strip()
```

```
def get_ansver_by_intent(intent):
  if intent in BOT_CONFIG['intents']:
      responses=BOT_CONFIG['intents'][intent]['responses']
  else:
      responses=['']
  return  random.choice(responses)
```

Используем датасет NLP_dataset с диалогами обмена фразами.Вероятность совпадения фраз с введеным текстом при большом обьеме датасета достаточно велика. 

```
with open('/content/dialogues.txt') as f:
  content=f.read()
dialogues_str=content.split('\n\n' ) #разделяем текст на строки диалога
dialogues=[dialogues_str.split('\n' )[:2] for dialogues_str in dialogues_str] #создаем слоаарь вопрос -ответ из двух фраз

dialogues_filter=[]
questions=set()                                   # используем  для выбора уникальных значений вопросов
for dialogue in dialogues:
  if len(dialogue) !=2:
    continue
  question,answer=dialogue                        # выбор иэ диалого вопрос - ответ
  question= clear_phrase(question[2:])            #Убираем из каждой вопроса 2 первых символа и проверем на вхождение каждого символа в словарь alfabet
  answer=answer[2:]                               #Убираем из каждой ответа 2 первых символа 
  if question !='' and question not in questions:
      questions.add(question) 
      dialogues_filter.append([question,answer])
#Генерируем структуру данных,состоящую из словаря слов в волрос -ответов  в которые данное слово входит
#Данную операцию будем осушествлять один раз при запуске и эта структура позволит быстрее искать реплики  с данным словом
dialogues_strucured={}
for question,ansver in dialogues_filter :
  words=set(question.split(' '))                    # разделяем слова
  for word in words:
    if word not in dialogues_strucured:             # если слово в первый раз
      dialogues_strucured[word]=[]                  # создадим пустую сруктуру 
    dialogues_strucured[word].append([question,ansver])#добавляем запись в структу

dialogues_strucured_cut={}
for word,pairs in dialogues_strucured.items():
  pairs.sort(key=lambda pair :len(pair[0]))   # сортируем по жлине фразы
  dialogues_strucured_cut[word]=pairs[:1000]  # обрезаем первые 1000 фраз       
```

```
import nltk
def generate_ansver(replica):
   replica=clear_phrase(replica)  
   words=set(replica.split(' ')) # разделяем слова в веденной реплике
   mini_dataset=[]               #Создаем мини сет волрос-ответ с данным словом 
   for word in words:
     if word  in dialogues_strucured_cut:
        mini_dataset+=dialogues_strucured_cut[word]
   ansvers=[] 
   for question,ansver in mini_dataset :
       if abs(len(replica)-len(question))/  len(question) <0.2:   #вычисляес разность длин реплики и ответа
           distance=nltk.edit_distance(replica, question)         #вычислям разницу между  репликой и ответом  растояние Левинштейна
           distance_weighted= distance / len(question)
           if distance_weighted <0.2:   
               ansvers.append([distance_weighted,question,ansver]) #создаем структуру [дистанция вопрос -ответ]

   if ansvers:
       return  min(ansvers,key=lambda tree:tree[0])[2]              #выбираем элемент с минимальной дистанцей
```

```
def get_failure_phrase():
  failure_phrase=BOT_CONFIG['failure_phrase']
  return random.choice(failure_phrase)
```

```
def bot(replica):
  #NLU
  intent=classify_intent(replica)
  #ansver generation
  #выбор заготовленной реплики
  if intent :
       ansver=get_ansver_by_intent(intent)
       if ansver:      
          stats['intent']+=1
          return ansver
  #вызов генеративной модели
  ansver=generate_ansver(replica)
  if ansver :
       stats['generate']+=1 
       return ansver

  #берем заглушку
  ansver=get_failure_phrase() 
  stats['failure']+=1       
  return ansver
```
